```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to the CSV file
file_path = '6.9.2024.csv'

# Read the CSV file
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.df()
```

```python
# Scatter plot for AVG vs HR
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(data['AVG'], data['HR'], color='blue')
plt.title('Batting Average vs Home Runs')
plt.xlabel('Batting Average (AVG)')
plt.ylabel('Home Runs (HR)')

# Scatter plot for AVG vs RBI
plt.subplot(1, 2, 2)
plt.scatter(data['AVG'], data['RBI'], color='green')
plt.title('Batting Average vs Runs Batted In')
plt.xlabel('Batting Average (AVG)')
plt.ylabel('Runs Batted In (RBI)')

# Show plots
plt.tight_layout()
plt.show()
```
