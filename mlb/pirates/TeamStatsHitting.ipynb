```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to print the dataframe
def print_dataframe(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)

    # Set display options to show all columns
    pd.set_option('display.max_columns', None)

    # Display the dataframe
    print(df)
    
# Function to plot batting averages
def plot_batting_averages(file_path):
    # Load the CSV file
    team_stats = pd.read_csv(file_path)

    # Filter relevant columns
    batting_averages = team_stats[['Tm', 'BA']]

    # Sort the batting averages from highest to lowest
    batting_averages_sorted = batting_averages.sort_values(by='BA', ascending=False)

    # Print the sorted batting averages
    print(batting_averages_sorted)

    # Define colors: black for Pittsburgh Pirates, yellow for others
    colors = ['black' if team == 'Pittsburgh Pirates' else 'yellow' for team in batting_averages_sorted['Tm']]

    # Plot the batting averages with the specified colors, y-axis formatting, and data labels
    plt.figure(figsize=(12, 8))
    bars = plt.bar(batting_averages_sorted['Tm'], batting_averages_sorted['BA'], color=colors)
    plt.xlabel('Teams')
    plt.ylabel('Batting Average')
    plt.title('Team Batting Averages (Sorted)')

    # Format the y-axis to show three decimal points
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.3f}'))

    # Add data labels centered on each bar
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.3f}', ha='center', va='bottom')

    # Adjust x-tick labels to be vertical and read better
    plt.xticks(rotation=90, ha='center')

    # Customize x-tick labels to be vertical and set padding
    plt.gca().tick_params(axis='x', which='both', labelbottom=True, bottom=True, top=False, pad=10)

    # Ensure layout is adjusted to show all x-ticks
    plt.tight_layout()
    plt.show()

# Function to plot home runs
def plot_home_runs(file_path):
    # Load the CSV file
    team_stats = pd.read_csv(file_path)

    # Remove the league sum row if it exists
    # Assuming the league sum row has a team name like 'League Total' or similar
    team_stats = team_stats[team_stats['Tm'] != 'League Sum']

    # Filter relevant columns
    home_runs = team_stats[['Tm', 'HR']]

    # Sort the home runs from highest to lowest
    home_runs_sorted = home_runs.sort_values(by='HR', ascending=False)

    # Define colors: black for Pittsburgh Pirates, yellow for others
    colors = ['black' if team == 'Pittsburgh Pirates' else 'yellow' for team in home_runs_sorted['Tm']]

    # Plot the home runs with the specified colors and data labels
    plt.figure(figsize=(12, 8))
    bars = plt.bar(home_runs_sorted['Tm'], home_runs_sorted['HR'], color=colors)
    plt.xlabel('Teams')
    plt.ylabel('Home Runs')
    plt.title('Team Home Runs (Sorted)')

    # Add data labels
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom')

    # Adjust x-tick labels to be vertical and read better
    plt.xticks(rotation=90, ha='center')

    # Customize x-tick labels to be vertical and set padding
    plt.gca().tick_params(axis='x', which='both', labelbottom=True, bottom=True, top=False, pad=10)

    # Ensure layout is adjusted to show all x-ticks
    plt.tight_layout()
    plt.show()

# File path to the CSV
file_path = 'TeamStatsHitting.csv'

# Execute functions
plot_batting_averages(file_path)
print_dataframe(file_path)
plot_home_runs(file_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot team stolen bases
def plot_stolen_bases(file_path):
    # Load the CSV file
    team_stats = pd.read_csv(file_path)

    # Remove the league sum row if it exists
    team_stats = team_stats[team_stats['Tm'] != 'League Sum']

    # Filter relevant columns
    stolen_bases = team_stats[['Tm', 'SB']]

    # Sort the stolen bases from highest to lowest
    stolen_bases_sorted = stolen_bases.sort_values(by='SB', ascending=False)

    # Print the sorted stolen bases
    print(stolen_bases_sorted)

    # Define colors: black for Pittsburgh Pirates, yellow for others
    colors = ['black' if team == 'Pittsburgh Pirates' else 'yellow' for team in stolen_bases_sorted['Tm']]

    # Plot the stolen bases with the specified colors and data labels
    plt.figure(figsize=(12, 8))
    bars = plt.bar(stolen_bases_sorted['Tm'], stolen_bases_sorted['SB'], color=colors)
    plt.xlabel('Teams')
    plt.ylabel('Stolen Bases')
    plt.title('Team Stolen Bases (Sorted)')

    # Add data labels centered on each bar
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom')

    # Adjust x-tick labels to be vertical and read better
    plt.xticks(rotation=90, ha='center')

    # Customize x-tick labels to be vertical and set padding
    plt.gca().tick_params(axis='x', which='both', labelbottom=True, bottom=True, top=False, pad=10)

    # Ensure layout is adjusted to show all x-ticks
    plt.tight_layout()
    plt.show()

# File path to the CSV
file_path = 'TeamStatsHitting.csv'

# Execute the function to plot stolen bases
plot_stolen_bases(file_path)
```
